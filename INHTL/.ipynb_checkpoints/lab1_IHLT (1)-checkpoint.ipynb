```python
import nltk
from string import punctuation
```

```python
pt = []
for x in punctuation:
  p = str(x)
  pt.append(p)
print(pt)
```

```python
nltk.download('gutenberg')
nltk.download('stopwords')
```

```python
txt = nltk.corpus.gutenberg.words('blake-poems.txt')
txt_lower = [x.lower() for x in txt]
```

```python
sw=set(nltk.corpus.stopwords.words('english'))
```

```python
txt_no_pt = []
for w in txt_lower:
  for ch in w.split():
    if ch in pt:
      break
    txt_no_pt.append(w)
print(txt_no_pt)
```

```python
freqs = {w:txt_no_pt.count(w) for w in set(txt_no_pt) if w not in sw}



s_freqs = sorted(freqs.items(), key=lambda x: x[1], reverse=True)
c = 0
for i in s_freqs:
  if c < 25:
    print(i[0], i[1])
    c = c + 1

```
