## Visualizar géneros de películas con MDS

```python
def MDS(D):                                                                                       
    # Number of points                                                                        
    n = len(D)
    # Centering matrix                                                                        
    H = np.eye(n) - np.ones((n, n))/n
    # YY^T                                                                                    
    B = -H.dot(D**2).dot(H)/2
    # Diagonalize                                                                             
    evals, evecs = np.linalg.eigh(B)
    # Sort by eigenvalue in descending order                                                  
    idx   = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:,idx]
    # Compute the coordinates using positive-eigenvalued components only                      
    w, = np.where(evals > 0)
    L  = np.diag(np.sqrt(evals[w]))
    V  = evecs[:,w]
    Y  = V.dot(L)
 
    return Y, evals
```

#### Definimos funciones para pintar los puntos en 2D y 3D

```python
from matplotlib.colors import to_rgb, to_rgba
import matplotlib.patches as mpatches
def scatter2D(x, y, color, alpha_arr, label, **kwarg):
    r, g, b = to_rgb(color)
    color_alpha = [(r, g, b, alpha / 5) for alpha in alpha_arr]
    plt.scatter(x, y, c = color_alpha, **kwarg)
    plt.xlabel("First dimension")
    plt.ylabel("Second dimension")
    plt.title("Metric MDS with Manhattan distance")
    plt.axis('equal');
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
def scatter3D(ax, x, y, z, color, alpha_arr, label, **kwarg):
    r, g, b = to_rgb(color)
    color_alpha = [(r, g, b, alpha / 5) for alpha in alpha_arr]
    ax.scatter(xs = x, 
           ys = y, 
           zs = z,
           color = color_alpha, **kwarg)
```
