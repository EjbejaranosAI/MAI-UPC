# [V] Clasificación

El objetivo de la clasificación es predecir la categoria $y$ a partir de un vector $d$-dimensional $x \in \mathbb{R}^d$ de variables de entrada.

Hay que distinguir entre clasificación binaria (2 clases) y clasificación multi-clase (más de 2 clases).

Hay diferentes algoritmos para hacer clasificación: hoy veremos tres, que son ...

- Regresión logística (aunque se llame regresión, es clasificación binária).
- k-NN (k Nearest Neighbors).
- SVM (Support Vector Machine).

---

### Regresión logística

La regresión logística sigue la fórmula:

$$\hat{y_i} = f(x_i,w) = \sigma\left(w_0 + w_1\phi_1(x) + w_2\phi_2(x) + ... + w_m\phi_m(x)\right)=\sigma\left(\sum_{j=0}^m w_j\phi_j(x)\right)$$

Donde $\phi$ es una función de base y $\sigma$ es la función logística, $\sigma(x)=\frac{1}{1+e^{-x}}$. ¿Cómo es capaz de clasificar? Pues porque la función sigmoide lleva a dos extremos (uno corresponde a la clase 0 y el otro corresponde a la clase 1):

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
X = np.arange(-10, 10, 0.1)
Y = sigmoid(X)

plt.plot(X, Y)
plt.xlabel('x')
plt.ylabel('y');
```

### k-NN (k Nearest Neighbors)

Se trata de clasificar el punto según la clase de los $k$ vecinos más cercanos.

- 1-NN se trata de clasificar el punto $x$ según el vecino más cercano, i.e. $argmin_{x'}d(x, x')$.
- Para k-NN consideramos $k\geq 1$ vecinos cercanos, si tenemos solo dos clases, para $k$ impar no podemos tener empates, pero en el resto de los casos si que pueden haber empates.
- Cuando hay empates, la mejor manera de desampatar es aleatoriamente ... (nadie ha demostrado una manera mejor).

A medida que aumentamos la $k$ ...

- Hay menos influencia del ruido (fluctuaciones locales de las clases).
- La variancia baja.
- Pero la localidad se pierde.
- Y el coste computacional crece.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Iris dataset
X = np.c_[load_iris().data[:, 0],load_iris().data[:, 2]]
Y = load_iris().target
```

```python
Y
```

```python
Xc = StandardScaler().fit_transform(X)
```

```python
from sklearn import neighbors

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
Xc_train, Xc_test, Yc_train, Yc_test = train_test_split(Xc, Y, test_size = 0.3, random_state = 0)

clf1 = neighbors.KNeighborsClassifier(n_neighbors = 1)
clf2 = neighbors.KNeighborsClassifier(n_neighbors = 1)

clf3 = neighbors.KNeighborsClassifier(n_neighbors = 3)
clf4 = neighbors.KNeighborsClassifier(n_neighbors = 3)

clf5 = neighbors.KNeighborsClassifier(n_neighbors = 20)
clf6 = neighbors.KNeighborsClassifier(n_neighbors = 20)

clf1.fit(X_train, Y_train)
clf2.fit(Xc_train, Yc_train)
clf3.fit(X_train, Y_train)
clf4.fit(Xc_train, Yc_train)
clf5.fit(X_train, Y_train)
clf6.fit(Xc_train, Yc_train)
```

```python
from matplotlib.colors import ListedColormap

precision = 0.1
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, precision),
                     np.arange(y_min, y_max, precision))

xc_min, xc_max = Xc_train[:, 0].min() - 1, Xc_train[:, 0].max() + 1
yc_min, yc_max = Xc_train[:, 1].min() - 1, Xc_train[:, 1].max() + 1
xxc, yyc = np.meshgrid(np.arange(xc_min, xc_max, precision),
                     np.arange(yc_min, yc_max, precision))

fig, ax = plt.subplots(3, 2, figsize = (10, 8))

for idx, clf, xx, yy, X, Y in zip([[0,0], [0,1], [1,0], [1,1], [2,0], [2,1]],
                    [clf1, clf2, clf3, clf4, clf5, clf6],
                    [xx, xxc, xx, xxc, xx, xxc],
                    [yy, yyc, yy, yyc, yy, yyc],
                    [X_train, Xc_train, X_train, Xc_train, X_train, Xc_train],
                    [Y_train, Yc_train, Y_train, Yc_train, Y_train, Yc_train]):
    Z = clf.predict(np.c_[xx.reshape(-1), yy.reshape(-1)])
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    ax[idx[0], idx[1]].pcolormesh(xx, yy, Z, cmap = cmap_light)

    # Plot also the training points
    ax[idx[0], idx[1]].scatter(X[:, 0], X[:, 1], c = Y, cmap = cmap_bold,
                edgecolor = 'k', s = 20)

    ax[idx[0], idx[1]].set_xlim(xx.min(), xx.max())
    ax[idx[0], idx[1]].set_ylim(yy.min(), yy.max())
    ax[idx[0], idx[1]].set_title("{}-NN clasificación (k = {})".format(clf.n_neighbors, clf.n_neighbors))

plt.show()
```

```python
clf1 = neighbors.KNeighborsClassifier(n_neighbors = 3, 
                                     metric = 'chebyshev')
clf2 = neighbors.KNeighborsClassifier(n_neighbors = 3, 
                                     metric = 'chebyshev')

clf3 = neighbors.KNeighborsClassifier(n_neighbors = 3, 
                                     metric = 'manhattan')
clf4 = neighbors.KNeighborsClassifier(n_neighbors = 3,
                                     metric = 'manhattan')

clf5 = neighbors.KNeighborsClassifier(n_neighbors = 3, 
                                     metric = 'euclidean')
clf6 = neighbors.KNeighborsClassifier(n_neighbors = 3,
                                     metric = 'euclidean')

clf1.fit(X_train, Y_train)
clf2.fit(Xc_train, Yc_train)
clf3.fit(X_train, Y_train)
clf4.fit(Xc_train, Yc_train)
clf5.fit(X_train, Y_train)
clf6.fit(Xc_train, Yc_train)

precision = 0.1
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, precision),
                     np.arange(y_min, y_max, precision))

xc_min, xc_max = Xc_train[:, 0].min() - 1, Xc_train[:, 0].max() + 1
yc_min, yc_max = Xc_train[:, 1].min() - 1, Xc_train[:, 1].max() + 1
xxc, yyc = np.meshgrid(np.arange(xc_min, xc_max, precision),
                     np.arange(yc_min, yc_max, precision))

fig, ax = plt.subplots(3, 2, figsize = (10, 8))

for idx, clf, xx, yy, X, Y in zip([[0,0], [0,1], [1,0], [1,1], [2,0], [2,1]],
                    [clf1, clf2, clf3, clf4, clf5, clf6],
                    [xx, xxc, xx, xxc, xx, xxc],
                    [yy, yyc, yy, yyc, yy, yyc],
                    [X_train, Xc_train, X_train, Xc_train, X_train, Xc_train],
                    [Y_train, Yc_train, Y_train, Yc_train, Y_train, Yc_train]):
    Z = clf.predict(np.c_[xx.reshape(-1), yy.reshape(-1)])
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    ax[idx[0], idx[1]].pcolormesh(xx, yy, Z, cmap = cmap_light)

    # Plot also the training points
    ax[idx[0], idx[1]].scatter(X[:, 0], X[:, 1], c = Y, cmap = cmap_bold,
                edgecolor = 'k', s = 20)

    ax[idx[0], idx[1]].set_xlim(xx.min(), xx.max())
    ax[idx[0], idx[1]].set_ylim(yy.min(), yy.max())
    plt.title("{}-NN clasificación (k = {})".format(clf.n_neighbors, clf.n_neighbors))

plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

X = np.c_[load_iris().data[:, 0],load_iris().data[:, 2]]
Y = load_iris().target
Xc = StandardScaler().fit_transform(X)

maxk = 30    #El oroginal estaba en 20

modelos = []
nombres = [str(k) + 'NN classifier' for k in range(1, maxk + 1)]
for k in range(1, maxk + 1): modelos.append(neighbors.KNeighborsClassifier(n_neighbors = k))
    
resultados_train = []
    
for nombre, modelo in zip(nombres, modelos):
    
    rkf = RepeatedKFold(n_splits = 5, n_repeats = 10, random_state = 123)
    resultados_nk = []
    
    for train_index, test_index in rkf.split(Xc):
        Xc_train_nk, Xc_test_nk = X[train_index], X[test_index]
        Yc_train_nk, Yc_test_nk = Y[train_index], Y[test_index]
        
        scores_train = cross_val_score(modelo, Xc_train_nk, Yc_train_nk, cv = 5, scoring = 'accuracy')
        resultados_nk.append(scores_train.mean())
        
        
    resultados_train.append(resultados_nk)
    print("%s: %f (%f)" % (nombre, np.mean(resultados_nk), np.std(resultados_nk)))
    
    print()
```

```python
import seaborn as sns
fig, ax = plt.subplots(1, 1, figsize = (10,5)) 
ax.set_title('Accuracy de kNN entre k = 1...20')
r = [np.mean(knn) for knn in resultados_train]
ax.plot(range(1, maxk + 1), r, label = 'train')
ax.vlines(np.argmax(r) + 1, np.min(r), 
          np.max(r),  
          color = 'gray', linestyle = 'dashed')
ax.legend()
print('Núm óptimo de vecinos: ', np.argmax(r) + 1)
```

### SVM (Support Vector Machine)

La base se fonamenta en los *métodos kernel*. SVC (*Support Vector Classifier*) es un modelo de aprendizaje supervisado que trata de encontrar el margen separador óptimo entre dos clases.  Otra vez tenemos un problema de optimización.

<div style="display:flex; justify-content: center;">
    <img src="data/svm1.png" alt="drawing" style="height:250px;"/>
</div>

El margen separador es un hiperplano, i.e., un subespacio vectorial de una dimensión menos que la dimensionalidad de los datos. Gráficamente, esto es:

- Si tenemos datos bidimensionales (dos ejes), el separador en un plano es una recta.
- Si tenemos datos tridimensionales (tres ejes), el separador en un espacio es un plano.
$\vdots$
- Si tenemos datos de dimensionalidad $N$ ($N$ variables), el separador es un hiperplano de dimensionalidad $N-1$.

<div style="display:flex; justify-content: center;">
    <img src="data/svm2.png" alt="drawing" style="height:250px;"/>
</div>

En la clasificación de clases linealmente separables no hay ningún problema: la SVC trata de encontrar un hiperplano que maximize la distancia entre dos puntos de clases diferentes más cercanos y asegura que cada punto está en el bando correspondiente a su clase.


En `sklearn` hay la función `make_blobs` que nos simplifica la creación de nubes de datos Gaussianos ...

```python
import matplotlib.pyplot as plt
from sklearn import datasets
X, Y = datasets.make_blobs(n_samples = 100, centers = 2, n_features = 2, random_state = 20)
plt.plot(X[:, 0][Y == 0], X[:, 1][Y == 0], 'g^')
plt.plot(X[:, 0][Y == 1], X[:, 1][Y == 1], 'bs')
plt.show()
```

Pero qué hacemos cuando tenemos clases linealmente no separables?

```python
import matplotlib.pyplot as plt
from sklearn import datasets
X, Y = datasets.make_blobs(n_samples = 100, centers = 2, n_features = 2, random_state = 0)
plt.plot(X[:, 0][Y == 0], X[:, 1][Y == 0], 'g^')
plt.plot(X[:, 0][Y == 1], X[:, 1][Y == 1], 'bs')
plt.show()
Y[Y == 0] = -1
```

En el caso simple, la SVC encontraba un margen máximo condicionado a que las dos clases son linealmente separables. Pero si no lo son (como la mayoría de los datasets), el problema de optimización no encuentra la solución. 

Las dos propuestas para afrontar este problema son:
- Soft Margin
- Kernel Trick

#### Soft Margin
Se trata de encontrar un hiperplano separador que tolere un cierto error de clasificación. El margen separador óptimo es aquel que maximiza la distancia entre las dos clases y minimiza la suma de los errores (puntos clasificados incorrectamente).

Se llama `Soft Margin`en contraste con el caso linealmente separable, donde el margen lleva el nombre de `Hard Margin`.

En el problema de optimización se añade un término más, la suma de los errores, que viene ponderada por un hiperparámetro `C` (como más grande sea `C`, penalizamos más el error de clasificación incorrecta).

```python
from sklearn import svm
from sklearn.svm import SVC
# Por defecto, C = 1
clf1 = svm.SVC(kernel = 'linear', C = 1)

# Menos penalización + más tolerancia a error (más "soft")
clf2 = svm.SVC(kernel = 'linear', C = 0.01)

# Más penalización + menos tolerancia a error (más "hard")
clf3 = svm.SVC(kernel = 'linear', C = 20)
```

```python
fignum = 1
for modelo in [clf1, clf2, clf3]:
    modelo.fit(X, Y)

    plt.figure(fignum, figsize = (5, 5))
    plt.clf()

    plt.scatter(modelo.support_vectors_[:, 0], modelo.support_vectors_[:, 1], s = 80,
                facecolors = 'none', zorder = 10, edgecolors = 'k')
    plt.scatter(X[:, 0], X[:, 1], c = Y, zorder = 10, cmap = plt.cm.Pastel1,
                edgecolors = 'k')

    plt.axis('tight')
    x_min = np.min(X[:, 0]) - 1
    x_max = np.max(X[:, 0]) + 1
    y_min = np.min(X[:, 1]) - 1
    y_max = np.max(X[:, 1]) + 1

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = modelo.decision_function(np.c_[XX.reshape(-1), YY.reshape(-1)])

    Z = Z.reshape(XX.shape)
    plt.figure(fignum, figsize = (4, 3))
    plt.pcolormesh(XX, YY, Z > 0, cmap = plt.cm.Pastel1)
    plt.contour(XX, YY, Z, colors = ['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    plt.xticks(())
    plt.yticks(())
    fignum = fignum + 1
    
plt.show()
```

#### Kernel trick
Se trata de encontrar un separador no lineal en el espacio de entrada. Es decir, se busca otro espacio (al que se accede mediante una transformación) donde los datos son linealmente separables y se aplica el SVC para buscar el hiperlano separador en ese espacio.

Estas transformaciones se llaman "kernels". Hay muchos tipos de kernels, algunos estan implementados en `scikit` por defecto, pero también podemos crear nuestros propios kernels y aplicarlo en el SVC.

```python
clf1 = svm.SVC(kernel = 'linear', C = 1)

clf2 = svm.SVC(kernel = 'poly', C = 1) 

clf3 = svm.SVC(kernel = 'rbf', C = 1) 
```

```python
fignum = 1
for modelo in [clf1, clf2, clf3]:
    modelo.fit(X, Y)

    plt.figure(fignum, figsize = (5, 5))
    plt.clf()

    plt.scatter(modelo.support_vectors_[:, 0], modelo.support_vectors_[:, 1], s = 80,
                facecolors = 'none', zorder = 10, edgecolors = 'k')
    plt.scatter(X[:, 0], X[:, 1], c = Y, zorder = 10, cmap = plt.cm.Pastel1,
                edgecolors = 'k')

    plt.axis('tight')
    x_min = np.min(X[:, 0]) - 1
    x_max = np.max(X[:, 0]) + 1
    y_min = np.min(X[:, 1]) - 1
    y_max = np.max(X[:, 1]) + 1

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = modelo.decision_function(np.c_[XX.reshape(-1), YY.reshape(-1)])

    Z = Z.reshape(XX.shape)
    plt.figure(fignum, figsize = (4, 3))
    plt.pcolormesh(XX, YY, Z > 0, cmap = plt.cm.Pastel1)
    plt.contour(XX, YY, Z, colors = ['k', 'k', 'k'], linestyles=['--', '-', '--'],
                levels=[-.5, 0, .5])

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    plt.xticks(())
    plt.yticks(())
    fignum = fignum + 1
    
plt.show()
```

Cada kernel puede aportar adicionalmente sus hiperparámetros propios, por ejemplo, el kernel polinómico tiene el grado del polinomio como hiperparámetro, o el kernel RBF tiene la `gamma`. Esto complica las búsqueda del mejor modelo.
