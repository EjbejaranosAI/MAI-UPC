### Escalamiento Multidimensional (MDS)

*Multi-Dimensional Scaling* conocido como MDS. Es una clase de técnicas de exploración y visualización de datos muy potente. A partir de distancias entre observaciones… **¿Podemos calcular las coordenadas de estas observaciones en un espacio N-dimensional?**

Antes, necesitamos calcular las distancias entre las observaciones. 

**¿CÓMO DEFINIMOS LA DISTANCIA ENTRE LAS OBSERVACIONES?**

IMPORTANTE: **¿distancia respecto qué variables?**

#### Similitud vs Distancia

Necesitamos una métrica para comparar cómo de similares (o distantes) son 2 observaciones:

- **Similarity (*similitud*)**: $S(x,y)$ crece como más parecidas sean. $S(x,y) = S(y,x)$, $S(x,y) > 0$
- **Dissimilarity (*distancia*)**: $D(x,y)$ crece como menos parecidas sean. $D(x,y) = D(y,x)$, $D(x,y) >= 0$, $D(x,x) = 0$

La distancia será una métrica iff 
  $D(x,y) <= D(x,z) + D(z,y)$     y si    $D(x,y) = 0$ si $x = y$ *(Desigualdad del triángulo)*


Hay muchos tipos de distancias … La más conocida es la **Euclidiana**:

$ \delta_{rs} = \sqrt{(x_r - x_s)^T (x_r - x_s)} = \sqrt{\sum_{i = 1}^p (x_{r_i} - x_{s_i})^2} $

```python
def euclidean(xr, xs):
    p = len(xr) # len(xr) = len(xs)
    return np.sqrt(sum( (xr[i] - xs[i])**2 for i in range(0, p)))
```

```python
xr = np.array([2, 3, 4])
xs = np.array([4, 2, 3])

euclidean(xr, xr), euclidean(xr, xs)
```

```python
def euclidean_vec(xr, xs):
    return np.sqrt((xr - xs) @ (xr - xs).T)
```

```python
euclidean_vec(xr, xr), euclidean_vec(xr, xs)
```

La distancia **Mahalanobis**, es una generalización de la Euclidiana, tiene en cuenta la matriz de covarianzas:

$ \delta_{rs} = \sqrt{(x_r - x_s)^T \Sigma^{-1} (x_r - x_s)} $

Si la matriz de covarianzas es la identitad, obtendremos la distancia Euclidiana:

$ \delta_{rs} = \sqrt{(x_r - x_s)^T I_d (x_r - x_s)} = \sqrt{(x_r - x_s)^T (x_r - x_s)} = \sqrt{||x_r - x_s||^2} = ||x_r - x_s||$


*Ejemplo*: considera dos variables, $x_0$ y $x_1$, que correlacionan perfectamente, per en direcciones contrarias:

```python
X = np.array([[0, 2], [1, 1], [2, 0]]).T
C = np.cov(X)
C
```

El elemento $C_{01}$ de la matriz muestra la correlación entre $x_0$ y $x_1$, que es negativa, ya que $C_{01} = -1$.

```python
def mahalanobis(xr, C, xs):
    return np.sqrt((xr - xs) @ np.linalg.pinv(C) @ (xr - xs))
```

```python
x1 = np.array([0, 2])
x2 = np.array([1, 1])
x3 = np.array([2, 0])
```

```python
mahalanobis(x1, C, x2), euclidean(x1, x2)
```

```python
mahalanobis(x1, C, x1), euclidean(x1, x1)
```

```python
mahalanobis(x1, C, x3), euclidean(x1, x3)
```

**¿Cómo se interpreta esta distancia?** Esta métrica nos devuelve la distancia sobre datos "decorrelacionados" (quita información reduandante), es decir, estandarizados (puedes pensar como pasados a las mismas unidades "ocultas", para que las diferentes variables contribuyan lo mismo). Si utilizamos la distancia Mahalanobis, no tenemos que normalizar los datos "crudos", en cambio con la distancia Euclidiana si que lo deberíamos hacer.

La distancia de **Manhattan** o del taxista (distancia $L_1$):

$ \delta_{rs} = \sum_{i = 1}^p |x_{r_i} - x_{s_i}| $

```python
def manhattan(xr, xs):
    p = len(xr) # len(xr) = len(xs)
    return sum( np.abs(xr[i] - xs[i]) for i in range(0, p))
```

```python
xr = np.array([2, 3.5, 4, 8.5, 10.5, -5, 6, 7])
xs = np.array([4, 2.5, 3, 7.75, 11.35, -4, 6, 8])

euclidean(xr, xs), manhattan(xr, xs)
```

```python
x1 = np.array([0, 2])
x2 = np.array([2, 0])

euclidean(x1, x2), manhattan(x1, x2)
```

Hay estudios que afirman que en datos de alta dimensionalidad es preferible utilizar la distancia $L_1$ en vez de $L_2$. **¿Por qué?**

- La distancia Euclidiana es la manera más común de calcular la distancia entre vectores. Observa que las diferencias notables se exageran y las diferencias pequeñas se ignoran (el cuadrado de números pequeños se hace más pequeño, y el cuadrado de números grandes se hace más grande). Por eso, si dos vectores son casi iguales, la distancia Euclidiana los verá muy similares (poco disimilares).
- La distancia Manhattan es similar a la Euclidiana en el sentido que la escala importa (hay que estandarizar los datos), pero es diferente a la Euclidiana en el sentido de que **NO IGNORA** las pequeñas diferencias. Si dos vectores son casi iguales, la distancia de Manhattan será más grande. 

Vamos a probar:

```python
xr = np.array([2.001, 3, 4.025])
xs = np.array([2, 2.998, 4.011])

euclidean(xr, xs), manhattan(xr, xs)
```

```python
xr = np.array([2, 5, 4])
xs = np.array([20000, 50000, 40000])

euclidean(xr, xs), manhattan(xr, xs)
```

La distancia de **Chebyshev** se define como:
    
$ \delta_{rs} = \max_i (|x_{r_i} - x_{s_i}|) $

```python
def chebyshev(xr, xs):
    return np.max(np.abs(xr - xs))
```

```python
xr = np.array([2, 3, 4])
xs = np.array([2, 349, 4])

euclidean(xr, xs), chebyshev(xr, xs)
```

```python
xr = np.array([2, 3.5, 4, 8.5, 10.5, -5, 6, 7])
xs = np.array([4, 2.5, 3, 7.75, 11.35, -4, 6, 8])
print(xr - xs)

euclidean(xr, xs), chebyshev(xr, xs)
```

La distancia de **Minkowski** generaliza las distancias Euclidiana, Manhattan y Chebyshev.

$ \delta_{rs} = (\sum_{i = 1}^p |x_{r_i} - x_{s_i}|^p)^{\frac{1}{p}} $

- Si $p = 2$, Euclidiana $L_2$.
- Si $p = 1$, Manhattana $L_1$.
- Si $p \rightarrow \infty$, Chebyshev $L_{\infty}$.

```python
def minkowski(xr, xs, k):
    p = len(xr) # len(xr) = len(xs)
    return (sum((np.abs(xr[i] - xs[i]))**k for i in range(0, p)))**(1/k)
```

```python
xr = np.array([2.001, 3, 4.025])
xs = np.array([2, 2.998, 4.011])

euclidean(xr, xs), minkowski(xr, xs, k = 2)
```

```python
manhattan(xr, xs), minkowski(xr, xs, k = 1)
```

```python
chebyshev(xr, xs), minkowski(xr, xs, k = 100)
```

```python
x1 = np.array([0,0])
x2 = np.array([1,2])

euclidean(x1, x2), manhattan(x1, x2), chebyshev(x1, x2)
```

La distancia de **Hamming** es para vectores binarios. Si 0 y 0 $\rightarrow$ 0, si 1 y 1 $\rightarrow$ 0, y si 0 y 1 $\rightarrow$ 1. Coincide con la distancia de Manhattan para vectores binarios.

```python
def hamming(xr, xs):
    return sum(np.abs(xr - xs))
```

```python
xr = np.array([1,0,0,0,0,1,1,1,0,1,0,1])
xs = np.array([1,0,0,1,0,1,0,1,0,1,0,0])

hamming(xr, xs), euclidean(xr, xs), manhattan(xr, xs)
```

#### Tipos de MDS

Básicamente hay 2:
- **Metric MDS**: la distancia en el mapa es una aproximación de la distancia real → Principal Coordinate Analysis (PCoA).
- **Non-metric MDS**: aproxima el rango, no el valor de la distancia.

#### PCoA (Principal Coordinate Analysis)

Hace una aproximación por low-rank approximation (https://en.wikipedia.org/wiki/Low-rank_approximation). No lo podríamos hacer sin autovalores y autovectores… Aproximamos la matriz de productos escalares de las observaciones $b_{rs} = x_r^T x_s$.

La matriz original: $B = VD_{\lambda}V^T = \sum_{i=1}^N \lambda_i v_i v_i^T$

La aproximación: $\hat{B} = V_{(,1:k)}D_{(1:k, 1:k)}V_{(,1:k)}^T$


Para medir cómo de bien estamos aproximando la matriz de distancias, hay que calcular el GOF (*Goodness-of-Fit*, bondad de ajuste):

$$\frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^N |\lambda_i|} \space ó \space \frac{\sum_{i=1}^k \lambda_i}{\sum_{\lambda_i > 0} \lambda_i}$$

Como más grande sea - mejor! 

La implementación de PCoA prestada http://www.nervouscomputer.com/hfs/cmdscale-in-python/. En Matlab y R la función viene incluida en una librería, en Python no hay.

```python
def MDS(D):                                                                                       
    # Number of points                                                                        
    n = len(D)
    # Centering matrix                                                                        
    H = np.eye(n) - np.ones((n, n))/n
    # YY^T                                                                                    
    B = -H.dot(D**2).dot(H)/2
    # Diagonalize                                                                             
    evals, evecs = np.linalg.eigh(B)
    # Sort by eigenvalue in descending order                                                  
    idx   = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:,idx]
    # Compute the coordinates using positive-eigenvalued components only                      
    w, = np.where(evals > 0)
    L  = np.diag(np.sqrt(evals[w]))
    V  = evecs[:,w]
    Y  = V.dot(L)
 
    return Y, evals
```

### Principal Component Analysis (PCA)

Es una técnica que ...
- Nos permite reducir el número de variables.
- Nos permite identificar factores (variables sintéticas) importantes en los datos que permiten describir los datos en pocas dimensiones.
- Funciona analizando la varianza de los datos.

Las nuevas variables se crean a partir de combinaciones lineales de las variables existentes.

Estas nuevas variables reciben el nombre de **Componentes Principales** (PC).

$$
PC_1 = a_{11}X_1 + a_{12}X_2 + ... + a_{1p}X_p \\
PC_2 = a_{21}X_1 + a_{22}X_2 + ... + a_{2p}X_p \\
\vdots \\ 
PC_p = a_{p1}X_1 + a_{p2}X_2 + ... + a_{pp}X_p 
$$

$p$ es el número total de variables y $a_{ij}$ son los pesos. Si $a_{ij} = 0$ significa que la variable que multiplica no contribuye a la Componente Principal. Podemos definir hasta $p$ PCs, si intentamos definir más, tendremos redundancia.

- El primer Componente Principal (PC1) es en la dirección de más variabilidad (covarianza) en los datos.
- El segundo Componente Principal (PC2): es el siguiente orthogonal (sin correlación) a la dirección del de más variabilidad
- Y así succesivamente.

#### Las nuevas variables

- Son combinaciones lineales de las originales.
- No tienen relación entre ellas: no estan correlacionadas, es decir, no comparten variancia, son independientes. Geométricamente, esto implica que ...
- Son ortogonales en el espacio original.
- Capturan tanta varianza como en los datos originales.
- La varianza máxima es capturada por PC1.
- $Var(PC_1) \geq Var(PC_2) \geq ... \geq Var(PC_p)$
- $a_{i1}^2 + a_{i2}^2 + ... + a_{ip}^2 = 1, (-1 \leq a_{ij} \leq 1)$

#### ¿Cómo se encuentran?

Se ha demostrado que los pesos óptimos son los valores propios de la matriz de covariancia. Todo los coeficientes y valores propios se pueden obtener a partir de la descomposición espectral de la matriz de covariancia:
$$C = V D_{\lambda} V^T$$
Y estimar las componentes con:
$$PC_{n \times p} = X_{n \times p} V_{p \times p}$$
$X$ es la matriz de los datos (debe estar centrada). Geométricamente esto corresponde a una rotación.

```python
np.random.seed(1)
X = (np.random.rand(2, 2) @ np.random.randn(2, 1000)).T
```

```python
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("X1")
plt.ylabel("X2")
plt.axis('equal');
```

Claramente hay una relación lineal entre $X_1$ y $X_2$.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) # Queremos 2 PCA componentes
pca.fit(X)                  # Ajustarlas a los datos con .fit
```

```python
# PC componentes
print(pca.components_)
```

$$
PC_1 = -0.95X_1 -0.31X_2 \\
PC_2 = 0.31X_1 -0.95X_2 
$$

```python
# Variancia explicada por las componentes
print(pca.explained_variance_)
```

```python
def draw_vector(v0, v1, ax = None):
    ax = ax or plt.gca()
    arrowprops = dict(arrowstyle = '->',
                    linewidth = 2,
                    shrinkA = 0, shrinkB = 0, color = "red")
    ax.annotate('', v1, v0, arrowprops = arrowprops)
```

```python
# plot data
plt.scatter(X[:, 0], X[:, 1], alpha=1)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    draw_vector(pca.mean_, pca.mean_ + v)
plt.axis('equal');
```

### Aplicaciones de PCA

#### (1) Reducción de dimensionalidad con PCA

Originalmente, tenemos dos variables $X_1$ y $X_2$. Vamos a intentar a reducir los datos a una solo dimensión:

```python
pca = PCA(n_components = 1) # Solo 1 PC
pca.fit(X)
X_pca = pca.transform(X) # Aplicar reducción de dimensionalidad a los datos
print("original shape:   ", X.shape)      # 200 points in 2D (original)
print("transformed shape:", X_pca.shape)  # 200 points in 1D (reducido)
```

```python
X_new = pca.inverse_transform(X_pca) # Obtener los datos reducidos
plt.scatter(X[:, 0], X[:, 1], alpha = 0.25)
plt.scatter(X_new[:, 0], X_new[:, 1], alpha = 0.75)
plt.axis('equal');
```

Aunque perdamos la información de la dispersión vertical de la nube de datos (PC2), con una solo dimensión conseguimos  explicar el 80% de variancia.

Claramente, podemos ver qué es lo que hace PCA: la información de los ejes PC menos importantes es eliminada - solo nos quedamos con las componentes con la variancia más elevada. La fracción de variancia que es eliminada corresponde aproximadamente a la cantitad de "información" que perdemos en la reducción de dimensionalidad.

#### (2) Visualización de datos 

Quizás no podamos asimilar la utilidad de PCA con datos bidimensionales. Vamos a dejarlo más claro utilizando datos de alta dimensionalidad. 

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

Los datos son imagenes de números de 8x8 píxels, lo cual significa que son datos de 64 dimensiones. Cada píxel lo interpretamos como una variable que puede ser o bien 0 (negro) o bien 1 (blanco). Para visualizar el dataset y las relaciones (patrones), podemos bajar la dimensionalidad de 64 a 2:

```python
pca = PCA(n_components = 2)  # Proyectar 64 dimensiones a 2
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
plt.scatter(projected[:, 0], projected[:, 1],
            c = digits.target, edgecolor = 'none', alpha = 0.9,
            cmap = plt.cm.get_cmap('Spectral', 10))
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar()
```

El dataset es un conjunto de puntos que forman una nube 64-dimensional, y los puntos pintados arriba son la proyección de cada uno de los puntos 64D sobre las direcciones de máxima variancia (PC1 y PC2). Los colores representan la etiqueta de los números.

Las Componentes Principales actuan como una represantación de baja dimensionalidad de los datos.

##### ¿Cuántas componentes?

Se puede determinar el número de componentes mirando al plot de variancia acumulada en función del número de componentes.

```python
pca = PCA().fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_)) # cumsum computes the cumulative sum
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

Podemos observar que las primeras 10 componentes contienen aproximadamente el 75% de variancia, mientras que necesitariamos unas 50 componentes para describir aproximadamente el 100% de variancia. Vemos que con una proyección bidimensional perdemos mucha información (mesurada como la variancia explicada) y que necesitariamos unas 20 componentes para explicar el 90% de variancia.

### (3) Filtrar ruido

PCA también se puede usar para filtrar datos ruidosos (con error). La idea es la siguiente: una parte de la variancia corresponde al ruido (variancia producida por errores). Esta variancia deberia corresponder a componentes de importancia baja, por eso, si reconstruimos los datos utilizando el subconjunto de Componentes Principales más importantes, de esta forma eliminaremos parte del ruido.

```python
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize=(10, 4),
                             subplot_kw = {'xticks':[], 'yticks':[]},
                             gridspec_kw = dict(hspace = 0.1, wspace = 0.1))
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8, 8),
                  cmap = 'binary', interpolation = 'nearest',
                  clim = (0, 16))
```

```python
plot_digits(digits.data) # Datos originales, sin ruido
```

Vamos a añadir un poco de ruido:

```python
np.random.seed(42)
noisy = np.random.normal(digits.data, 4)
plot_digits(noisy)
```

Claramente, vemos que las imagenes son ruidosas, contienen píxeles erroneos. Vamos a ajustar PCA en estos datos ruidoso, preservando el 50% de variancia:

```python
pca = PCA(0.50).fit(noisy)
pca.n_components_
```

Preservamos el 50% de variancia de los datos originales con 12 Componentes Principales. Ahora, calculamos estas componentes, y usamos la inversa de la transformación para reconstruir los dígitos filtrados:

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

Vamos a poner más error aún, imaginemos que enviamos los dígitos por un canal muy ruidoso. Los números son casi incomprensibles. 

```python
np.random.seed(42)
noisy = np.random.normal(digits.data, 8)
plot_digits(noisy)
```

```python
pca = PCA(0.32).fit(noisy)
pca.n_components_
```

```python
components = pca.transform(noisy)
filtered = pca.inverse_transform(components)
plot_digits(filtered)
```

### Biplots

Función de pintar biplots: https://stackoverflow.com/questions/39216897/plot-pca-loadings-and-loading-in-biplot-in-sklearn-like-rs-autoplot

Biplot es una herramienta potente de exploración de datos para detectar patrones, anomalias, grupos y outliers. Pinta los ejes de las variables originales sobre el plot de proyección por PCA. Los ejes tienden a ser oblicuos, si el ángulo es pequeño entre las variables, significa que estan más correlacionadas. Si se forma un ángulo de 180, tenemos correlación negativa. El origen del biplot representa la media de las variables.

```python
def biplot(score, coeff, colors, labels = None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0 / (xs.max() - xs.min())
    scaley = 1.0 / (ys.max() - ys.min())
    plt.figure(figsize = (10,6))
    plt.scatter(xs * scalex,ys * scaley, c = colors)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var" + str(i + 1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

#Call the function. Use only the 2 PCs.
biplot(projected[:, 0:2], np.transpose(pca.components_[0:2, :]), Y, X.columns)
plt.show()
```
