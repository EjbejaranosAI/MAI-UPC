# [VI] Clustering

El objetivo de clustering es particionar los datos en grupos (conocidos como "clusters") de manera que las observaciones que se encuentran en el mismo cluster tienden a ser m치s similares que observaciones en clusters diferentes.

Clustering nos permite ...

- Aplicando clustering sobre observaciones - encontrar grupos "naturales" o la estructura taxon칩mica de los datos.
- Tambi칠n es una forma de reducir la dimensionalidad si aplicamos clustering sobre variables (agrupar variables en "supervariables").

Hay diferentes maneras de "clusterizar" (agrupar) y se han formado varias fam칤lias de algoritmos de clustering.

- Probabilistas (k-Means, E-M, ...)
- Algor칤tmicos (PAM, ...)
- Jerarquicos
- [...]

Clustering es aprendizaje no supervisado ya que no hay etiquetas $Y$, son los clusters que crean la etiqueta.

https://scikit-learn.org/stable/modules/clustering.html

Dos ingredientes para clustering:

- El tipo de distancia a aplicar.
- El tipo de algoritmo a aplicar.

## k-Means

Posiblemente el m치s conocido. Antes de todo - debemos definir cu치ntos clusters queremos encontrar (el # de clusters es un hiperpar치metro 游땩). El algoritmo es simple:

```
Inicializar aleatoriamente los asignaciones 
Repetir:
    (re)calcular los centroides
    (re)calcular las asignaciones
Hasta converger: no cambios en asignaciones o m치x # iteraciones
```


    <img src="data/cluster2.png" alt="drawing" style="height:500px;"/>
</div>

En cada momento asignamos un punto al cluster del centroide m치s cercano. La funci칩n que optimizamos es pues la distancia de cada punto respecto los $K$ centroides:

$$
Loss(c_1, ..., c_K) = \sum_{n=1}^N \sum_{k=1}^K I_{nk} ||x_n - c_k||^2
$$

Donde:

- $x_n$ es un dato de los $N$ que hay ($\sum_{n=1}^N$ por cada dato ...)
- $c_k$ es un centroide de los $K$ que hay ($\sum_{k=1}^K$ por cada centroide ...)
- $I_{nk}$ es una indicadora que indica si $x_n$ est치 asignada al cluster $K$.
- Minimizamos la distancia entre $x_n$ y $c_k$. 
- Queremos que los clusters sean compactos internamente y distantes entre ellos.

Ahora mates: si queremos el centroide 칩ptimo, tenemos que derivar respecto el centroide e igualar a cero:

$$
\frac{\partial Loss(c_1, ..., c_K)}{\partial c_k}= 2 \sum_{n=1}^N  I_{nk} (x_n - c_k) = 0
$$

Y nos sale que el centroide 칩ptimo es:
$$
c_{k_{opt}} = \frac{\sum_{n=1}^N I_{nk} x_n}{ \sum_{n=1}^N I_{nk}}
$$

- $\sum_{n=1}^N I_{nk} x_n$ es la suma de los puntos asignados al cluster $k$.
- $\sum_{n=1}^N I_{nk}$ es el n칰mero de puntos asignados al cluster $k$ (suma de indicadoras).

Es decir, el centroide 칩ptimo es la media de los puntos asignados al cluster $k$!!!

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Ejemplo que vimos ya un d칤a ...
x1 = np.random.multivariate_normal([0,4], np.eye(2), 100)
x2 = np.random.multivariate_normal([5,8], np.eye(2), 100)

plt.scatter(x1[:,0], x1[:,1], label = 'Cluster 1')
plt.scatter(x2[:,0], x2[:,1], label = 'Cluster 2', marker = 'x')
plt.legend()
````

````python
# Juntamos las filas de forma que ya vimos un d칤a ...
datos = pd.DataFrame(np.vstack((x1, x2)),  columns = ['X1','X2'])
````

````python
# N칰mero de clusters que queremos encontrar
K = 4

np.random.seed(123)

# X1 coordenadas de centroides aleatorios
C_X1 = np.random.randint(0, np.max(datos.X1), size = K)

# X2 coordenadas de centroides aleatorios
C_X2 = np.random.randint(0, np.max(datos.X2), size = K)

C = np.array(list(zip(C_X1, C_X2)), dtype = np.float32)
print(C)
````

````python
plt.scatter(datos.X1, datos.X2, c = 'black', s = 10)
plt.scatter(C_X1, C_X2, marker = '*', c = 'red', s = 200);
````

````python
# Funci칩n que definimos un d칤a ...
def euclidean(xr, xs):
    p = len(xr) # len(xr) = len(xs)
    return np.sqrt(sum( (xr[i] - xs[i])**2 for i in range(0, p)))
````

````python
def error_kMeans(datos, C, K, clusters):
    loss = 0
    for i in range(0, K):
        points = np.array([datos.loc[j] for j in range(len(datos.X1)) if clusters[j] == i])
        for j in range(0, len(points)):
            loss += euclidean(points[j], C[i])
    return loss  
````

````python
error_kMeans(datos, C, K, np.zeros(len(datos.X1)))
````

````python
def kMeans(datos, K, maxit = 10):
    """
        INPUT
        datos - en los que queremos encontrar clusters
        K - K clusters a encontrar
        
        OUTPUT
        clusters - asignaci칩n a cluster de cada observaci칩n
        C - centroides 
        
    """
    it = 0
    
    # X1 coordenadas de centroides aleatorios
    C_X1 = np.random.randint(0, np.max(datos.X1), size = K)

    # X2 coordenadas de centroides aleatorios
    C_X2 = np.random.randint(0, np.max(datos.X2), size = K)

    # Guardamos los centroides en una variable
    C = np.array(list(zip(C_X1, C_X2)), dtype = np.float32)

    # Etiquetas de los puntos
    clusters = np.zeros(len(datos.X1))

    # Error func
    error_ = 0
    error = error_kMeans(datos, C, K, clusters)
    
    # Hasta converger
    while np.abs(np.linalg.norm(error_) - np.linalg.norm(error))> 0.0001 and it <= maxit:
        print("It[{}] error: {}".format(it, np.linalg.norm(error)))
        
        # Asignar cada dato al cluster del centroide m치s cercano
        for i in range(0, len(datos.X1)):

            # Distancia entre dato y clusters
            distances = [euclidean(datos.loc[i], centroide) for centroide in C]

            # Escogemos la distancia m치s peque침a (el cluster del centroide m치s cercano)
            cluster = np.argmin(distances)
            
            # Asignamos este dato a este cluster
            clusters[i] = cluster
        
        # Calcular los nuevos centroides
        for i in range(0, K):
            points = np.array([datos.loc[j] for j in range(len(datos.X1)) if clusters[j] == i])
            if len(points) == 0: continue
            C[i] = np.mean(points, axis = 0)
        
        error_ = error
        error = error_kMeans(datos, C, K, clusters)
        it = it + 1
    print("It[*] error: {}".format(np.linalg.norm(error)))
    return clusters, C
````

````python
# N칰mero de clusters que queremos encontrar
K = 2

np.random.seed(100)

clusters, C = kMeans(datos = datos, K = K, maxit = 100)
colors = ['indianred', 'seagreen', 'cornflowerblue', 'gold', 'plum', 'orchid']
fig, ax = plt.subplots()
for i in range(K):
        points = np.array([datos.loc[j] for j in range(len(datos.X1)) if clusters[j] == i])
        if len(points) == 0: continue
        ax.scatter(points[:, 0], points[:, 1], s = 10, c = colors[i], label = "cluster {}".format(i + 1))
ax.scatter(C[:, 0], C[:, 1], marker = '*', s = 200, c = 'black')
plt.legend()
plt.show()
print(clusters)
````

## Clustering jer치rquico "agglomerative"
Se trata de comenzar por peque침os clusters (inicialmente cada punto es un cluster) y vamos uniendo los clusters para hacer grupos m치s grandes. Se unen siempre clusters que tienen m칤nimo distancia entre ellos.

#### Ventaja vs k-Means
- No tenemos el hiperpar치metro $K$!! Podemos definir tantos clusters como queramos despu칠s de crear el dendograma.

#### Desventaja vs k-Means
- Puede ser demasiado lento para datasets grandes.

Otra observaci칩n:
- Las observaciones no pueden cambiar de grupo.

<div style="display:flex; justify-content: center;">
    <img src="data/tree.png" alt="drawing" style="height:300px;"/>
</div>

Antes hemos dicho "distancia entre clusters". 쯈u칠 es eso? No es distancia entre dos puntos, si no distancia entre dos grupos de puntos. Hay diferentes estrategias para calcular esta distancia, a continuaci칩n presentamos una serie de estas:

- **Single Linkage (o Nearest Neighbor)**: la distancia entre dos clusters es la distancia m치s peque침a entre dos puntos, cada uno perteneciendo a un cluster diferente: $D(C_1, C_2) = \min(d(x_{C_1}, x_{C_2}))$
- **Complete Linkage (o Farthest Neighbor)**: la distancia entre dos clusters es la mayor distancia entre dos puntos, cada uno perteneciendo a un cluster diferente: $D(C_1, C_2) = \max(d(x_{C_1}, x_{C_2}))$
- **Average Linkage**: la distancia entre dos clusters es la distancia media entre cada punto de un cluster y otro: $D(C_1, C_2)= \frac{1}{N_{C_1}N_{C_2}}\sum_{i=1}^{N_{C_1}}\sum_{j=1}^{N_{C_2}}d(x_{{C_1}_i}, x_{{C_2}_j})$
- **Centroid**: la distancia entre dos clusters es la distancia entre los centroies de ambos.

Hay m치s ... (por ej. **ward**, que calcula el grado de dispersi칩n de las observaciones del cluster respecto todas las variables).

````python
from scipy.spatial.distance import pdist, squareform
dissimilarities = squareform(pdist(X, 'cityblock')) # Manhattan
````

````python
def MDS(D):                                                                                       
    # Number of points                                                                        
    n = len(D)
    # Centering matrix                                                                        
    H = np.eye(n) - np.ones((n, n))/n
    # YY^T                                                                                    
    B = -H.dot(D**2).dot(H)/2
    # Diagonalize                                                                             
    evals, evecs = np.linalg.eigh(B)
    # Sort by eigenvalue in descending order                                                  
    idx   = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:,idx]
    # Compute the coordinates using positive-eigenvalued components only                      
    w, = np.where(evals > 0)
    L  = np.diag(np.sqrt(evals[w]))
    V  = evecs[:,w]
    Y  = V.dot(L)
 
    return Y, evals
````

````python
coords, eigenvals = MDS(dissimilarities)
````

````python
plt.scatter(x = coords[:, 0], y = coords[:, 1], c = labels)
plt.xlabel("First dimension")
plt.ylabel("Second dimension")
plt.title("Metric MDS with Manhattan distance")
plt.axis('equal');
````

Hm... 쮺칩mo decido que distancia de clustering escoger?

Algunas pistas:

- **Single Linkage** a침ade muy tarde a los outliers, por eso tiene la tendencia de indentificar clusters en forma de cadena. Sensible a outliers.
- **Complete Linkage** a침ade rapidamente a los outliers. Sensible a outliers.
- **Average Linkage** y **Centroid**: son menos sensibles a outliers.
- **Ward**: menos sensible a outliers y tiende a crear clusters de tama침o similar.
