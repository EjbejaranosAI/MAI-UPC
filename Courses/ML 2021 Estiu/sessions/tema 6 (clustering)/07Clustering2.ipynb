### Calinksi-Harabasz score

También conocido "Variance Ratio Criterion" o "Pseudo F Index", es una métrica que indica el ratio entre la variancia entre clusters respecto la variancia intra-cluster. Valores altos de Pseudo F indican clusters compactos y separados entre si, en particular, buscamos picos en la gráfica de número de clusters VS Pseudo F Index.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.random.multivariate_normal([0,4], np.eye(2), 100)
x2 = np.random.multivariate_normal([5,8], np.eye(2), 100)
X = pd.DataFrame(np.vstack((x1, x2)),  columns = ['X1','X2'])
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics

indexF = []
for K in range(2, 21):
    kmeans_model = KMeans(n_clusters = K, random_state = 1).fit(X)
    labels = kmeans_model.labels_
    indexF.append(metrics.calinski_harabasz_score(X, labels))
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
ax[0].plot(range(2, 21), indexF)
ax[1].scatter(x1[:,0], x1[:,1], label = 'Cluster 1')
ax[1].scatter(x2[:,0], x2[:,1], label = 'Cluster 2', marker = 'x')
plt.legend()
```

```python
from sklearn.datasets import make_blobs

N = 1000
randstate = 170

X, Y = make_blobs(n_samples = N, centers = 3, n_features = 2, random_state = randstate)

indexF = []
for K in range(2, 21):
    kmeans_model = KMeans(n_clusters = K, random_state = 1).fit(X)
    labels = kmeans_model.labels_
    indexF.append(metrics.calinski_harabasz_score(X, labels))
    
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
ax[0].plot(range(2, 21), indexF)
ax[1].scatter(X[:,0], X[:,1]);
```

### Silhouette index

El índice de la silueta es otra métrica para evaluar la calidad del clustering. Este índice estima la distancia media entre clusters y como más alto sea el valor del índice de la silueta, mejor están definidos los clusters.

```python
from sklearn.metrics import silhouette_score

N = 1000
randstate = 170

X, Y = make_blobs(n_samples = N, centers = 3, n_features = 2, random_state = randstate)

silueta = []
for K in range(2, 21):
    kmeans_model = KMeans(n_clusters = K, random_state = 1).fit(X)
    labels = kmeans_model.labels_
    silueta.append(metrics.silhouette_score(X, labels))
    
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
ax[0].plot(range(2, 21), silueta)
ax[1].scatter(X[:,0], X[:,1]);
```

### Davies-Bouldin Index

Un valor bajo del índice de Davies-Bouldin indica una mejor separación entre clusters. El valor más bajo posible es el 0. El índice se interpreta como una medida de "similitud" media entre clusters, donde esta similitud se calcula comparando la distancia entre clusters y el tamaño de los propios clusters.

```python
from sklearn.metrics import silhouette_score

N = 1000
randstate = 170

X, Y = make_blobs(n_samples = N, centers = 3, n_features = 2, random_state = randstate)

dbindex = []
for K in range(2, 21):
    kmeans_model = KMeans(n_clusters = K, random_state = 1).fit(X)
    labels = kmeans_model.labels_
    dbindex.append(metrics.davies_bouldin_score(X, labels))
    
fig, ax = plt.subplots(1, 2, figsize = (15, 5))
ax[0].plot(range(2, 21), dbindex)
ax[1].scatter(X[:,0], X[:,1]);
```
